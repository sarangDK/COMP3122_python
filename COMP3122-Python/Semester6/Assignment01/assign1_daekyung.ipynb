```python
# Question 1 - import essentail libraries
from keras.datasets import fashion_mnist
import numpy as np
import matplotlib.pyplot as plt


# Load the dataset
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
# Question 10 - Please import models and layers from the keras library
from keras.models import Sequential
from keras.layers import Dense, Flatten
```

```python
# Question 11 - Define a sequential model and call it myNetwork
myNetwork = Sequential()
```

```python
# Question 12, 13 - Reshape the images from 28x28 to one column with 784 neurons (flattening)
# Also, please normalize the image by dividing the image by 255
myNetwork.add(Flatten(input_shape=(28, 28)))
```

```python
# Question 14 - Add one hidden layer that has 512 neurons, using the ‘relu’ activation function.
myNetwork.add(Dense(512, activation='relu'))

```

```python
# Question 15 - # Question 15: Add another hidden layer with 128 neurons and ReLU activation

myNetwork.add(Dense(128, activation='relu'))
```

```python
# Question 16 - Add the last layer as a 10-neuron dense layer that uses the ‘softmax’ as the
# activation function. Why do we use softmax for the last layer? How does it work
# under the hood? Make sure to explain this in your video

myNetwork.add(Dense(10, activation='softmax'))
```

```python
# Question 17 A - Use the following two settings for the compiler and run them separately and see what the differences are.
myNetwork.compile(optimizer='adam', loss='sparse_categorical_crossentropy',metrics=['accuracy'])

# Question 17 B
# myNetwork.compile(optimizer='rmsprop',
#                   loss='categorical_crossentropy',
#                   metrics=['accuracy'])
```

```python
# Question 18 -  Now after the compilation, please try to find the pattern using the fit command. Set the number of epochs to 10
myNetwork.fit(train_images, train_labels, epochs=10)
```

```python
# Question 21 - Use the evaluate() function from the Keras library to calculate the achieved accuracy
# and loss over the test images and labels. Do we have overfitting? Explain in your video
test_loss, test_accuracy = myNetwork.evaluate(test_images / 255.0, test_labels)
print(f"Test Accuracy: {test_accuracy}")
print(f"Test Loss: {test_loss}")



# If training accuracy is much higher than test accuracy, overfitting might be present.
```
