```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
age = np.loadtxt("biostats.csv", delimiter=",", skiprows=1, usecols=2)
```

```python
age_array = np.array(age)
age_array_content = age_array
print(age_array_content)
```

```python
average = np.mean(age_array_content)
standard_deviation = np.std(age_array_content)
min_age = np.min(age_array_content)
max_age = np.max(age_array_content)


print(f"Average of the age is : {average}")
print(f"Standard deviation of the age is : {standard_deviation}")
print(f"Minimum age is : {min_age}")
print(f"Maximum age is : {max_age}")

```

```python
percentile_30 = np.percentile(age_array_content, 30)
median_age = np.median(age_array_content)
percentile_70 = np.percentile(age_array_content, 70)

print(f"30th percentile of the age is : {percentile_30}")
print(f"Median of the age is : {median_age}")
print(f"70th percentile of the age is : {percentile_70}")
```

```python
plt.hist(age_array_content, bins=10, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Number')
plt.title('Age Distribution - By Daekyung')
plt.show()
```
